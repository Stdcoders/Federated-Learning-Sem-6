```python
import numpy as np
```

```python
class LinearModel:
    def __init__(self):
        self.w = np.random.randn()
        self.b = np.random.randn()
```

```python
class Client:
    def __init__(self, client_id, X, y):
        self.client_id = client_id
        self.X = X
        self.y = y
        self.model = LinearModel()

    def set_global_model(self, global_model):
        self.model.w = global_model.w
        self.model.b = global_model.b

    def train(self, lr=0.01, epochs=50):
        w, b = self.model.w, self.model.b

        for _ in range(epochs):
            y_pred = w * self.X + b
            dw = np.mean((y_pred - self.y) * self.X)
            db = np.mean(y_pred - self.y)

            w -= lr * dw
            b -= lr * db

        self.model.w = w
        self.model.b = b

    def get_weights(self):
        return self.model.w, self.model.b
```

```python
class Server:
    def __init__(self):
        self.global_model = LinearModel()

    def send_model(self, clients):
        for client in clients:
            client.set_global_model(self.global_model)

    def federated_averaging(self, client_weights):
        avg_w = np.mean([w for w, b in client_weights])
        avg_b = np.mean([b for w, b in client_weights])

        self.global_model.w = avg_w
        self.global_model.b = avg_b

    def get_global_model(self):
        return self.global_model
```

```python
def federated_learning(num_clients=5, rounds=5):
    server = Server()

    clients = []
    for i in range(num_clients):
        X = np.random.rand(50) * 10
        y = 2 * X + 3 + np.random.randn(50)
        clients.append(Client(i, X, y))

    for r in range(rounds):
        print(f"Round {r+1}")

        server.send_model(clients)

        client_weights = []
        for client in clients:
            client.train()
            client_weights.append(client.get_weights())

        server.federated_averaging(client_weights)

    return server.get_global_model()
```

```python
final_model = federated_learning()
print("Final Global Model:")
print("w =", final_model.w)
print("b =", final_model.b)
```
