```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Student_Performance.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
label_encoder = LabelEncoder()
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
X = df.drop(["final_grade", "student_id"], axis=1)
y = df["final_grade"]
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled.head()
```

```python
processed_df = pd.concat([X_scaled, y.reset_index(drop=True)], axis=1)
processed_df.head()
```

```python
train_data, test_data = train_test_split(
    processed_df,
    test_size=0.2,
    random_state=42
)
```

```python
num_clients = 4
client_datasets = np.array_split(train_data, num_clients)
```

```python
for i, client_data in enumerate(client_datasets):
    print(f"Client {i+1} dataset shape: {client_data.shape}")
```
